```python
import pandas as pd
import wandb_api as wapi

wapi.init_wandb_api()

GROUP = 'kw14-test-functionality'
```

```python
runs = wapi.get_runs_by_regex('fftower-conv-imagenet1k-lr0.01-frag[1-8]', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Frags (lr: 0.01)', label_attr='batch_frag')
```

```python
runs = wapi.get_runs_by_regex('fftower-conv-imagenet1k-lr0.001-frag[1-8]', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Frags (lr: 0.001, Adam)', label_attr='batch_frag')
```

```python
runs = wapi.get_runs_by_regex('fftower-conv-imagenet1k-lr0.0001-frag[1-8]', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Frags (lr: 0.0001, Adam)', label_attr='batch_frag')
```

```python
runs = wapi.get_runs_by_regex('fftower-conv-imagenet1k-lr0.001-frag[1-8]-sgd', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Frags (lr: 0.001, SGD)', label_attr='batch_frag')
```

```python
runs = wapi.get_runs_by_regex('fftower-[a-z]*-imagenet1k-lr0.0001-ts1$', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Cell Types (lr: 0.0001, Adam, T=1)', label_attr='cell_type')
```

```python
runs = wapi.get_runs_by_regex('fftower-[a-z]*-lr0.0001-ts1-sgd$', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Cell Types (lr: 0.0001, SGD, T=1)', label_attr='cell_type')
```

```python
runs = wapi.get_runs_by_regex('fftower-rnn-lr0.0001-ts[0-9]*-adam$', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Enrollments (lr: 0.0001, Adam, RNN)', label_attr='time_steps')
```

```python
runs = wapi.get_runs_by_regex('fftower-rnn-lr0.0001-ts[0-9]*-sgd$', group=GROUP)
wapi.plot_runs_accuracy(runs,'Different Enrollments (lr: 0.0001, SGD, RNN)', label_attr='time_steps')
```
